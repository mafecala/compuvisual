```python
# Importar las bibliotecas necesarias
import trimesh
import numpy as np
import os
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import random

# Configurar el entorno para visualización en Jupyter
%matplotlib inline

# Definir la ruta al archivo OBJ (una carpeta arriba del directorio actual)
ruta_archivo = os.path.join('..', 'magnolia.obj')

# Cargar el modelo 3D
modelo = trimesh.load(ruta_archivo)
print(f"Modelo '{ruta_archivo}' cargado exitosamente!")

# Función para generar colores distintos para cada malla
def generar_color():
    return (random.random(), random.random(), random.random())

# Variables para acumular información global
total_vertices = 0
total_caras = 0
total_aristas = 0
todas_mallas = []

# Extraer todas las mallas si es una escena, o usar la malla directamente
if isinstance(modelo, trimesh.Scene):
    print("El modelo cargado es una escena que contiene múltiples mallas.")
    # Extraer todas las mallas de la escena
    mallas = list(modelo.geometry.values())
    print(f"La escena contiene {len(mallas)} mallas.")
    
    # Acumular información de todas las mallas
    for i, malla in enumerate(mallas):
        todas_mallas.append(malla)
        total_vertices += len(malla.vertices)
        total_caras += len(malla.faces)
        total_aristas += len(malla.edges)
        print(f"Malla {i+1}: {len(malla.vertices)} vértices, {len(malla.faces)} caras, {len(malla.edges)} aristas")
else:
    # Si es una malla individual, usarla directamente
    print("El modelo cargado es una malla individual.")
    malla = modelo
    todas_mallas.append(malla)
    total_vertices = len(malla.vertices)
    total_caras = len(malla.faces)
    total_aristas = len(malla.edges)

# Mostrar información estructural global del modelo
print("\n--- INFORMACIÓN ESTRUCTURAL DEL MODELO COMPLETO ---")
print(f"Número total de vértices: {total_vertices}")
print(f"Número total de caras: {total_caras}")
print(f"Número total de aristas: {total_aristas}")
print(f"Número de mallas: {len(todas_mallas)}")
print("--------------------------------------------------")

# Crear figura para visualizar todas las mallas 3D
fig = plt.figure(figsize=(14, 12))
ax = fig.add_subplot(111, projection='3d')

# Variables para determinar los límites globales del gráfico
todos_vertices = []

# Visualizar cada malla con un color distinto para caras
for idx, malla in enumerate(todas_mallas):
    # Generar colores para esta malla
    color_vertices = 'red'
    color_aristas = 'blue'
    color_caras = generar_color()  # Color aleatorio para cada malla
    
    vertices = malla.vertices
    todos_vertices.extend(vertices)
    
    # Visualizar vértices (puntos)
    ax.scatter(vertices[:, 0], vertices[:, 1], vertices[:, 2], 
               color=color_vertices, s=5, alpha=0.5)
    
    # Visualizar aristas (líneas)
    for inicio, fin in malla.edges_unique:
        # Extraer coordenadas de los vértices que forman la arista
        x = [vertices[inicio, 0], vertices[fin, 0]]
        y = [vertices[inicio, 1], vertices[fin, 1]]
        z = [vertices[inicio, 2], vertices[fin, 2]]
        # Dibujar la línea que representa la arista
        ax.plot(x, y, z, color=color_aristas, linewidth=0.3, alpha=0.7)
    
    # Visualizar caras (triángulos)
    triangulos = malla.faces
    x = vertices[:, 0]
    y = vertices[:, 1]
    z = vertices[:, 2]
    
    # Crear una representación de malla con caras transparentes para ver estructura
    ax.plot_trisurf(x, y, z, triangles=triangulos, alpha=0.2, color=color_caras)

# Convertir lista de vertices en array para cálculos
todos_vertices = np.array(todos_vertices)

# Configurar los límites de los ejes para mejor visualización
# Calcular el centro y la escala para centrar y ajustar la visualización
if len(todos_vertices) > 0:
    centro = np.mean(todos_vertices, axis=0)
    escala = np.max(np.abs(todos_vertices - centro))
    
    # Establecer límites de ejes
    ax.set_xlim([centro[0] - escala, centro[0] + escala])
    ax.set_ylim([centro[1] - escala, centro[1] + escala])
    ax.set_zlim([centro[2] - escala, centro[2] + escala])

# Etiquetas de ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Título del gráfico
ax.set_title('Visualización 3D del modelo completo: magnolia.obj')

# Crear leyenda personalizada
import matplotlib.patches as mpatches
rojo = mpatches.Patch(color='red', label='Vértices')
azul = mpatches.Patch(color='blue', label='Aristas')
verde = mpatches.Patch(color='green', label='Caras (colores varían por malla)')
ax.legend(handles=[rojo, azul, verde])

# Mostrar la figura
plt.tight_layout()
plt.show()

# Información adicional sobre el modelo
print("\n--- INFORMACIÓN ADICIONAL DEL MODELO ---")
total_volumen = 0
total_area = 0
mallas_watertight = 0

for i, malla in enumerate(todas_mallas):
    print(f"\nMalla {i+1}:")
    print(f"  Volumen: {malla.volume:.6f} unidades cúbicas")
    print(f"  Área de superficie: {malla.area:.6f} unidades cuadradas")
    print(f"  ¿Es una variedad (manifold)?: {malla.is_watertight}")
    print(f"  Centro de masa: {malla.center_mass}")
    
    total_volumen += malla.volume
    total_area += malla.area
    if malla.is_watertight:
        mallas_watertight += 1

print("\nTotales:")
print(f"  Volumen total: {total_volumen:.6f} unidades cúbicas")
print(f"  Área total: {total_area:.6f} unidades cuadradas")
print(f"  Mallas watertight: {mallas_watertight} de {len(todas_mallas)}")
print("----------------------------------------")

# Visualizar normales de las caras para cada malla
fig = plt.figure(figsize=(14, 12))
ax = fig.add_subplot(111, projection='3d')

for idx, malla in enumerate(todas_mallas):
    if hasattr(malla, 'face_normals'):
        # Obtener los centros de las caras para dibujar las normales
        centros_caras = np.mean(malla.vertices[malla.faces], axis=1)
        
        # Dibujar una muestra de normales para no saturar la visualización
        muestra = range(0, len(centros_caras), max(1, len(centros_caras) // max(50, 500 // len(todas_mallas))))
        
        for i in muestra:
            inicio = centros_caras[i]
            fin = inicio + malla.face_normals[i] * (escala * 0.05)  # Escalar las normales
            x = [inicio[0], fin[0]]
            y = [inicio[1], fin[1]]
            z = [inicio[2], fin[2]]
            ax.plot(x, y, z, color=generar_color(), linewidth=0.7)

# Configurar los límites de los ejes para la visualización de normales
ax.set_xlim([centro[0] - escala, centro[0] + escala])
ax.set_ylim([centro[1] - escala, centro[1] + escala])
ax.set_zlim([centro[2] - escala, centro[2] + escala])

# Título y etiquetas
ax.set_title('Normales de las caras de todas las mallas')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.tight_layout()
plt.show()
```
